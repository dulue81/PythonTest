```python
import numpy as np
import tensorflow as tf
import pandas as pd
pd.options.mode.chained_assignment = None
import seaborn as sns
from matplotlib.pylab import rcParams
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go

%matplotlib inline

sns.set(style='whitegrid', palette='muted')
rcParams['figure.figsize'] = 14, 8
np.random.seed(1)
tf.random.set_seed(1)

print('Tensorflow version:', tf.__version__)
```

```python
df = pd.read_csv('/Users/derek/Documents/GitHub/PythonTest/spx.csv')
```

```python
df
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df.date,y=df.close,mode='lines',name='close'))
fig.update_layout(showlegend=True)
fig.show()
```

```python
train_size = int(len(df)*0.8)
test_size =  int(len(df)*0.2)
train, test = df.iloc[0:train_size], df.iloc[train_size:len(df)]
train.shape, test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler = scaler.fit(train[['close']])
train['close'] = scaler.transform(train[['close']])
test['close'] = scaler.transform(test[['close']])
```

```python
train.describe()
```

```python
test.describe()
```

```python
def create_sequences(X, y, time_steps = 1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X.iloc[i:(i+time_steps)].values)
        ys.append(y.iloc[i+time_steps])
    return np.array(Xs), np.array(ys)
```

```python
time_steps = 30
X_train, y_train = create_sequences(train[['close']], train.close, time_steps)
X_test, y_test = create_sequences(test[['close']], test.close, time_steps)
X_train.shape, y_train.shape
```

```python
X_train
```

# Build an LSTM Autoencoder

```python
n_features = X_train.shape[2]
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout, RepeatVector, TimeDistributed
```

```python
model = Sequential()
model.add(LSTM(128,input_shape=(X_train.shape[1:])))
model.add(Dropout(rate = 0.2))
model.add(RepeatVector(X_train.shape[1]))
model.add(LSTM(128,return_sequences=True))
model.add(Dropout(rate=0.2))
model.add(TimeDistributed(Dense(n_features)))
model.summary()
```

```python
opt = keras.optimizers.Adam(learning_rate=1e-3)
model.compile(optimizer = 'adam',
             loss = 'mae',
             )
```

```python
len(X_train)*0.9
```

```python
es = keras.callbacks.EarlyStopping(monitor = 'val_loss', patience=3,mode='min')
history = model.fit(
    X_train, y_train,
    epochs = 100,
    batch_size = 32,
    validation_split = 0.1,
    callbacks = [es],
    shuffle = False
)
```

```python
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
plt.show()
```

```python
# get train mae loss:
X_train_pred = model.predict(X_train)
train_mae_loss = np.mean(np.abs(X_train_pred - X_train), axis=1)

plt.hist(train_mae_loss, bins=50)
plt.xlabel("Train MAE loss")
plt.ylabel("No of samples")
plt.show()

# Get reconstruction loss threshold.
threshold = np.max(train_mae_loss)
print("Reconstruction error threshold: ", threshold)
```

```python
# Get test MAE loss.
X_test_pred = model.predict(X_test)
test_mae_loss = np.mean(np.abs(X_test_pred - X_test), axis=1)
test_mae_loss = test_mae_loss.reshape((-1))

plt.hist(test_mae_loss, bins=50)
plt.xlabel("test MAE loss")
plt.ylabel("No of samples")
plt.show()

# Detect all the samples which are anomalies.
anomalies = test_mae_loss > threshold
print("Number of anomaly samples: ", np.sum(anomalies))
print("Indices of anomaly samples: ", np.where(anomalies))
```

```python
test_score = pd.DataFrame(test[time_steps:])
test_score['loss'] = test_mae_loss
test_score['threshold'] = threshold
test_score['anomaly'] = test_score.loss > test_score.threshold
test_score['close'] = test[time_steps:].close
```

```python
test_score
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=test[time_steps:].date, y=test_score.loss,mode='lines',name='Test Loss'))
fig.add_trace(go.Scatter(x=test[time_steps:].date, y=test_score.threshold,mode='lines',name='threshold'))
fig.update_layout(showlegend=True)
fig.show()
```

```python
anomalies = test_score[test_score.anomaly == True]
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=test[time_steps:].date, y =scaler.inverse_transform(test[time_steps:].close),mode='lines',name='Close price'))
fig.add_trace(go.Scatter(x=anomalies.date, y =scaler.inverse_transform(anomalies.close),mode='markers',name='Anomaly'))
fig.update_layout(showlegend=True)
fig.show()
```
